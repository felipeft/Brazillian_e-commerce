Distribuição de valores de preços e fretes

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
orders_table["price"].hist(bins=50)
plt.title("Distribuição dos Preços")
plt.xlabel("Preço")
plt.ylabel("Frequência")
plt.show()

plt.figure(figsize=(10,5))
orders_table["freight_value"].hist(bins=50)
plt.title("Distribuição do Frete")
plt.xlabel("Frete")
plt.ylabel("Frequência")
plt.show()
```

Quantidade de pedidos por mês abarcando o periodo

```python
orders_table.set_index("order_purchase_timestamp").resample("M")["order_id"].count().plot(figsize=(12,5))
plt.title("Pedidos por Mês")
plt.ylabel("Nº de pedidos")
plt.show()
```
